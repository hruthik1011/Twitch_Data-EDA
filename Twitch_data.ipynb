##Twitch data Exploratory Data Analysis

DATA CLEANING

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('datasetV2.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.dropna()  #drop rows with missing values
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')  # Standardize column names
```

```python
df.columns
```

```python
#convert date columns
if 'stream_date' in df.columns:
    df['stream_date'] = pd.to_datetime(df['stream_date'])
```

```python
print("data cleaned successfully")
df.head()
```

```python
df.columns
```

##Univariate Analysis

stream Categories

```python
# Top 10 categories by viewers
top_categories = df.groupby('most_streamed_game')['total_views'].sum().sort_values(ascending=False).head(10)
```

```python
top_categories
```

Followers Distribution

```python
# Histogram of follower counts
plt.figure(figsize=(8, 6))
sns.histplot(df['total_followers'], bins=20, kde=True, color='orange')
plt.title('Distribution of Followers')
plt.xlabel('Number of Followers')
plt.ylabel('Frequency')
plt.show()
```

##Bivariate Analysis

Followers vs viewers

```python
# Scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='total_followers', y='total_views', data=df, alpha=0.6)
plt.title('Followers vs. Viewers')
plt.xlabel('Followers')
plt.ylabel('Viewers')
plt.show()
```

Correlation Matrix

```python
# Compute correlation
corr_matrix = df[['total_views', 'total_followers', 'total_games_streamed']].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

##Time-Series Analysis

Viewership Trends

```python
# Group by date and sum viewers
if 'stream_date' in df.columns:
    time_series = df.groupby('stream_date')['total_views'].sum()

    # Plot
    plt.figure(figsize=(10, 6))
    time_series.plot(color='purple')
    plt.title('Viewership Over Time')
    plt.xlabel('Date')
    plt.ylabel('Total Viewers')
    plt.show()
```

##Peak Streaming Hours (if data includes timestamps)

```python
if 'stream_time' in df.columns:
    # Extract hour from stream time
    df['hour'] = pd.to_datetime(df['stream_time']).dt.hour

    # Viewership by hour
    peak_hours = df.groupby('hour')['viewers'].mean()

    # Plot
    plt.figure(figsize=(8, 6))
    peak_hours.plot(kind='bar', color='green')
    plt.title('Average Viewers by Hour')
    plt.xlabel('Hour of the Day')
    plt.ylabel('Average Viewers')
    plt.show()
```

```python
if 'language' in df.columns:
    # Top regions by total viewers
    region_viewers = df.groupby('language')['total_views'].sum().sort_values(ascending=False)

    # Plot
    plt.figure(figsize=(8, 6))
    region_viewers.plot(kind='bar', color='salmon')
    plt.title('Viewership by language')
    plt.xlabel('language')
    plt.ylabel('Total Viewers')
    plt.xticks(rotation=45)
    plt.show()
```
